```python
# import numpy
import numpy as np 
print(np.__version__)  #to check version
print(np.__doc__)   #to print docs
# convert list to array
l = [1,23,4,5,6, "Anand"]
print(np.array(l))
arr1 = np.array([[1, 2, 3], [4, 5, 6]])  #it checks the dimention of an array
arr1
# To create matrix in python
print(np.matrix([1,2,3,4,5]))
# creation of 3-D array
arr2 = np.array([[[1,2] , [3,4] , [5,6]]])
arr2
# More ways to create an array
l = [1,2,3]
np.asarray(l)
np.asanyarray(l)
# tuple into array
t = ([1, 2, 3], [4, 5, 6])
np.array(t)
# To create function into an array 
# create a 3D matrix 
arr3 = np.fromfunction(lambda i, j : i == j ,  (3,3))
arr3
arr4 = np.fromfunction(lambda i , j : i * j , (5,5))
arr4
# Array from iterables 
iterator = (i for i in range(5))
arr5 = np.fromiter(iterator, float)
arr5

# np.arrange : to generate the sequesnce of a number
arr6 = np.arange(1, 10, 0.1) #bu default the space is 1
arr6
# linespace
arr7 = np.linspace(1,5,10)
arr7
```
